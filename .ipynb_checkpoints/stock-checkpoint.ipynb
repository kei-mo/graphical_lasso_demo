# Japan Stock Market

```python
import numpy as np
import pandas as pd 
import os
from sklearn.covariance import GraphicalLasso
from matplotlib import pyplot as plt 
```

```python
dataset = pd.read_csv("stockprice2019.csv",index_col=0)
data = np.array(dataset)
feature_names = dataset.columns
```

```python
data
```

## Standarization

```python
data_standard = (data - np.mean(data,axis=0)) / np.std(data,axis=0)
```

## Graph Lasso

```python
alpha = 0.4 # param for 
model = GraphicalLasso(alpha=0.4,verbose=True) 
model.fit(data_standard)

cov_mat = model.covariance_
pre_mat = model.precision_
```

```python
plt.imshow(cov_mat)
```
